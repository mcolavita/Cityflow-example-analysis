```python
#let's focus on the vehicle n flow_3_0

#restart the engine
eng.reset()
eng.next_step()

#now i want to create a Pandas dataframe and put in this dataframe all the info of this vehicle
#after that i'll create a csv file to save this informations

vehicleInfo = pd.DataFrame() # empty at the start


vehicleId = []
for i in range(182):
    vehicleId.append('flow_3_0')
vehicleInfo['Id'] = vehicleId

steps = []
for i in range(182):
        steps.append(i)
vehicleInfo['Step'] = steps

#chosen attributes: distance, running,derivable and speed
#create this 3 lists

distance=[]
running= []
speed = []
drivable = []

distance.append(eng.get_vehicle_info('flow_3_0')['distance'])
running.append(eng.get_vehicle_info('flow_3_0')['running'])
speed.append(eng.get_vehicle_info('flow_3_0')['speed'])
drivable.append(eng.get_vehicle_info('flow_3_0')['drivable'])
```

```python
vehicleInfo
```

```python
for step in range(1,182):
    eng.next_step()
    distance.append(eng.get_vehicle_info('flow_3_0')['distance'])
    running.append(eng.get_vehicle_info('flow_3_0')['running'])
    speed.append(eng.get_vehicle_info('flow_3_0')['speed'])
    drivable.append(eng.get_vehicle_info('flow_3_0')['drivable'])
```

```python
vehicleInfo['Running'] = running
vehicleInfo['Drivable'] = drivable
vehicleInfo['Distance'] = distance
vehicleInfo['Speed'] = speed
```
